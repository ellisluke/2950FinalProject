```python
## Chart volatility

chartV = []
for week in chartWeeks:
    songsThatWeek = shortChart.loc[shortChart.WeekID == week]
    chartV.append(songsThatWeek['Weeks on Chart'].sum())

volatileTable = pd.DataFrame(columns=['Date', 'ChartVScore'])
volatileTable.Date = chartWeeks
volatileTable.ChartVScore = chartV
volatileTable.head()
```

```python

def urlBuild(song, artist):
    url = "http://www.songlyrics.com/"
    aChunks = artist.replace(' ', '-')
    sChunks = song.split(' ')
    # for ch in aChunks:
        # url = url + ch
        # url = url + '-'
    url = url + aChunks
    url = url + '/'
    for ch in sChunks:
        url = url + ch
        url = url + '-'
    url = url + 'lyrics'
    return url
    
urlBuild("Big Shot", "Kendrick Lamar & Travis Scott")
    
```

```python
from bs4 import BeautifulSoup
import requests

failureStr = "error, no lyrics found"
def webGrab(url):
    html = requests.get(url, allow_redirects=True).text
    soup = BeautifulSoup(html, 'html5lib')
    try:
        lyrics = soup.find(id="songLyricsDiv").text
    except:
        return failureStr
    lyrics = lyrics.replace('\n', ' ')
    lyrics = lyrics.replace(',', '')
    lyrics = lyrics.replace('-', ' ')
    lyrics = lyrics.replace('!', '')
    lyrics = lyrics.replace('(', '')
    lyrics = lyrics.replace(')', '')
    lyrics = lyrics.replace('?', '')
    lyrics = lyrics.lower()
    return lyrics

```

```python
## BUILD URLS
lyricCharts = peakCharts
print(lyricCharts.shape[0])
lyricCharts['lyricLink'] = ""
lyricCharts['lyrics'] = ""

for index, row in lyricCharts.iterrows():
    lyricCharts.loc[index, 'lyricLink'] = urlBuild(row.Song, row.Performer)
```

```python
## TEST DF WEBGRAB
from IPython.display import clear_output
count = 0
failCount = 0
total = lyricCharts.shape[0]
redirectE = "Redirect error"

for index, row in lyricCharts.iterrows():
    count = count + 1
    try: 
        lyricCharts.loc[index, 'lyrics'] = webGrab(lyricCharts.loc[index, 'lyricLink'])
    except:
        lyricCharts.loc[index, 'lyricBool'] = False
        failCount = failCount + 1
        lyricCharts.loc[index, 'lyrics'] = redirectE
        print("Caught an error")
        print("failure number: ", failCount)
        continue
    if failureStr in lyricCharts.loc[index, 'lyrics'] or "sorry we have no" in lyricCharts.loc[index, 'lyrics']:
        failCount = failCount + 1
        lyricCharts.loc[index, 'lyricBool'] = False
        clear_output(wait=True)
        print("failure number: ", failCount)
    else:
        clear_output(wait=True)
        print("success")
        lyricCharts.loc[index, 'lyricBool'] = True
    print("total done: ", count)
    print("progress: ", count / total)
        
```
